```python
from PIL import Image
import os
```

```python
def delete_small_images(root_directory):
    for directory, subdirectories, files in os.walk(root_directory):
        for filename in files:
            if (
                filename.endswith(".jpg")
                or filename.endswith(".jpeg")
                or filename.endswith(".png")
            ):
                img_path = os.path.join(directory, filename)
                with Image.open(img_path) as img:
                    width, height = img.size
                if width < 350 or height < 350:
                    # Close the image file before deleting it
                    img.close()
                    os.remove(img_path)
                    print(
                        f"Deleted image {filename} as its resolution was below 350x350."
                    )


# Replace 'path_to_your_images' with the path to your image directory
delete_small_images("data/images")
```
