## Count Images Dataset

### 1. Raw Images Dataset

```python
import os
import matplotlib.pyplot as plt
```

```python
def count_files_in_folder(folder_path):
    file_count = 0
    for _, _, files in os.walk(folder_path):
        file_count += len(files)
    return file_count


def count_images_in_classes(dataset_path):
    class_counts = {}
    sum_file = 0
    for root, dirs, _ in os.walk(dataset_path):
        for directory in dirs:
            class_path = os.path.join(root, directory)
            images = [
                f
                for f in os.listdir(class_path)
                if f.endswith(".jpg") or f.endswith(".jpeg") or f.endswith(".png")
            ]
            class_counts[directory] = len(images)
            sum_file += len(images)
    return class_counts, sum_file


def plot_bar_chart(class_counts, title):
    plt.bar(class_counts.keys(), class_counts.values())
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Number of Images")
    plt.xticks(rotation=90)
    plt.show()
```

### 1.1 Phone Dataset

```python
dataset_path = r"../../Phone Dataset"
class_counts, sum_file = count_images_in_classes(dataset_path)
print("=============== Phone Dataset ===============")
print("Number of images in each class:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}")
print(f"\nTotal: {sum_file}")
plot_bar_chart(class_counts, "Number of Images in Each Class")
```
